**Average Ticket price of each Pclass, calculated by all pasengers, dead passengers, and survived passengers ?**

```python
# YOUR CODE HERE
all_psg = df.groupby('Pclass')[['Fare']].mean().rename(columns={'Fare':'all_Passengers_avg_fare'})
dead_psg = df[df['Survived']==0].groupby('Pclass')[['Fare']].mean().rename(columns={'Fare':'dead_Passengers_avg_fare'})
survived_psg = df[df['Survived']==1].groupby('Pclass')[['Fare']].mean().rename(columns={'Fare':'survived_Passengers_avg_fare'})
result = pd.concat([all_psg, dead_psg, survived_psg], axis = 1 )
result
```

**Drop id Column**

```python
# YOUR CODE HERE
ex1.drop('id', axis = 1)
```

**Change the column 'release_date' data type to datetime and create column named 'datetime_format'**



```python
# YOUR CODE HERE
ex1['datetime_format'] = pd.to_datetime(ex1['release_date'])
ex1
```

**Sort values by date from descending**

```python
# YOUR CODE HERE
ex1.sort_values('datetime_format', ascending=False)
```

**Drop release_date column**

```python
# YOUR CODE HERE
ex1.drop(['release_date'], axis=1, inplace=True)
ex1
```

**Convert Numeric Columns to Integer**


**Remove Dollar Signs**

```python
# YOUR CODE HERE
cols = ['production_budget', 'domestic_gross', 'worldwide_gross']
for col in cols:
  ex1[col] = ex1[col].map(lambda x: x.replace('$',''))

ex1.head()
```

**Remove Commas**

```python
# YOUR CODE HERE
for col in cols:
  ex1[col] = [x.replace(',','') if ',' in x else x for x in ex1[col]]

ex1.head()
```

**Convert Columns to Integer D-Type**

```python
# YOUR CODE HERE
for col in cols:
  ex1[col] = ex1[col].astype(int)

ex1.dtypes
```
